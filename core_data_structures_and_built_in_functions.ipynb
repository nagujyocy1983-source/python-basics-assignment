Craete a tuple of cities

```python
cities = ("Delhi", "Mumbai", "Hyderabad", "Chennai", "Kolkata")

city1, city2, city3, city4, city5 = cities

print(city1)
print(city2)
print(city3)
print(city4)
print(city5)

try:
    cities[0] = "Pune"
except TypeError:
    print("Tuples cannot be changed!")
```

Task 2 : Sets

```python
# Task 2 - Sets

python_passers = {"Alice", "Bob", "Charlie", "Diana"}
sql_passers = {"Charlie", "Diana", "Frank", "Grace"}

print(python_passers | sql_passers)
print(python_passers & sql_passers)
print(python_passers - sql_passers)
```
